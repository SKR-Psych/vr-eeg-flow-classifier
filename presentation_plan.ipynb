## Slide 3: The Psychological Target: Flow State

### Slide Information
- **Slide Title:** Why 'Flow' is the Therapeutic Target
- **Objective:** Introduce Csikszentmihalyi's flow theory and explain its neuro-rehabilitative benefits.

### Slide Content
- **What is Flow?** A state of optimal challenge-skill balance, intense focus, sense of control, and time distortion.
- **Why Flow in PD Rehab?**
  - *Focused Attention:* Helps patients ignore external distractions and sustain motor engagement.
  - *Challenge-Skill Balance:* Keeps the task in the "optimal challenge zone"—avoiding boredom (task too easy) and anxiety/frustration (task too hard).
  - *Reduced Self-Consciousness:* Decreases anxiety surrounding motor deficits, making therapeutic movement feel natural rather than clinical.
- **The Target:** Build a closed-loop system that continuously monitors and modulates the user's flow state.

### Graphical Recommendations

## Slide 4: Paradigm Shift: Adaptation vs. Symbiosis

### Slide Information
- **Slide Title:** Paradigm Shift: From Adaptation to Neuro-Symbiosis
- **Objective:** Clarify the core philosophical contribution of your PhD report (defining NSAI).

### Slide Content
- **Conventional Adaptation:**
  - Adjusting simple, surface-level task parameters (e.g., speed, spacing) based on predefined rules.
  - The algorithm remains structurally static; the user merely drives inputs.
- **Neuro-Symbiosis (NSAI):**
  - A closed-loop computational architecture where human brain activity and AI co-evolve.
  - The AI undergoes **structural modifications** (e.g., updating network topology, processing pathways, and decision thresholds) dynamically based on human neurophysiology.
  - Operates across multiple timescales: immediate control (seconds) and symbiotic evolution (sessions).

### Graphical Recommendations
- **Graphic:** Two contrasting loop diagrams side-by-side:
  - *Left:* Simple linear feedback loop (User Input $\rightarrow$ Static Controller $\rightarrow$ VR Parameter change).
  - *Right:* Entangled, bi-directional symbiotic loop (User Brain Oscillations $\leftrightarrow$ Evolving Neural Network $\leftrightarrow$ Dynamic VR Space).

## Slide 7: Goofi-Pipe Integration & Neural Biomarkers

### Slide Information
- **Slide Title:** Real-Time Signal Processing & Flow Biomarkers
- **Objective:** Detail the exact signal processing pipeline and neurophysiological features used.

### Slide Content
- **Preprocessing Pipeline:** 
  - $0.5 - 45\text{ Hz}$ zero-phase bandpass filter.
  - Independent Component Analysis (ICA) targeting eye blink components for artifact rejection.
  - Bad channel interpolation via spherical splines.
- **Target EEG Biomarkers:**
  - **Relative Band Powers:** Frontal Theta ($4-8\text{ Hz}$) indicating mental engagement, and Parietal Alpha ($8-12\text{ Hz}$) indicating motor automaticity.
  - **Prefrontal Shannon Entropy:** Measures signal complexity. Focus/flow states show decreased frontal signal complexity (lower entropy).
  - **Frontal-Parietal Phase Locking Value (PLV):** Measures phase synchronization between executive (frontal) and motor (parietal) regions in theta/alpha bands.

## Slide 9: Results: Subject-Dependent Calibration vs. Global

### Slide Information
- **Slide Title:** SDC Breakthrough & The Data Leakage Trap
- **Objective:** Detail your machine learning results and the critical scientific finding of temporal data leakage.

### Slide Content
- **The Global Model Failure:** Training a universal model across subjects (using Leave-One-Subject-Out GroupKFold) gives **~51% accuracy** (chance). Global models fail due to high inter-subject EEG variance.
- **Subject-Dependent Calibration (SDC) Solution:** Training individual classifiers for each subject. Compares RBF Support Vector Machines (SVM) and Random Forest (RF) with class-weight balancing.
- **The Data Leakage Trap (Thesis Highlight):**
  - *Random Shuffle Split:* Yields **75% - 80% accuracy** due to temporal autocorrelation leakage between overlapping adjacent windows.
  - *Chronological Split:* Evaluates the model on future, unseen trials. Yields a true, honest baseline of **51.57% mean balanced accuracy** across all subjects.
- **Conclusion:** SDC with chronological splits is mathematically honest and represents the realistic state-of-the-art starting point for closed-loop calibration.

### Graphical Recommendations
- **Graphic:** A comparison bar chart showing: 
  - Global model (GroupKFold): ~51.0%
  - SDC (Chronological split - true test): ~51.57%
  - SDC (Random shuffle - leakage): ~78.0% (labeled "TEMPORAL LEAKAGE TRAP - INFLATED ACCURACY" with a warning icon).
